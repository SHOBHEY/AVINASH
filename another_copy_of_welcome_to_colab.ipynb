# **Q1. Data Visualization and Statistical Measures:**

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the Iris dataset
iris_data = load_iris()

# Convert the data to a pandas DataFrame
iris_df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)


# Display the first few rows of the dataset
print(iris_df.head())

# Data Visualization
# Histograms
iris_df.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()

# Boxplots
plt.figure(figsize=(10, 8))
sns.boxplot(data=iris_df)
plt.show()

# Scatter plots
sns.pairplot(iris_df, kind='scatter')
plt.show()

# Statistical Measures
statistical_measures = iris_df.describe()
print(statistical_measures)
```

# **Q2. Row Echelon Form:**

```
import numpy as np

# To Set the random seed using the last two digits of my roll number i.e. 09
np.random.seed(9)

A = np.random.randint(0, 10, size=(5, 5))

def reduce_to_row_echelon_form(matrix):
    num_rows, num_cols = matrix.shape
    lead = 0
    for r in range(num_rows):
        if lead >= num_cols:
            break
        i = r
        while matrix[i, lead] == 0:
            i += 1
            if i == num_rows:
                i = r
                lead += 1
                if num_cols == lead:
                    return matrix
        matrix[[i, r]] = matrix[[r, i]]
        if matrix[r, lead] != 0:
            matrix[r] = matrix[r] / matrix[r, lead]
        for i in range(num_rows):
            if i != r:
                matrix[i] = matrix[i] - matrix[i, lead] * matrix[r]
        lead += 1
    return matrix

row_echelon_A = reduce_to_row_echelon_form(A)
print("Original matrix A:")
print(A)
print("\nRow Echelon Form of matrix A:")
print(row_echelon_A)
```

# **Q3. Singular Value Decomposition (SVD):**

```
import numpy as np

np.random.seed(22)
A = np.random.randint(0, 10, size=(5, 5))

# Perform Singular Value Decomposition (SVD) on matrix A
U, Sigma, Vt = np.linalg.svd(A)
print("Matrix U:")
print(U)
print("\nMatrix Sigma (Diagonal matrix):")
print(np.diag(Sigma))
print("\nMatrix Vt (Transpose of V):")
print(Vt)
A_reconstructed = np.dot(U, np.dot(np.diag(Sigma), Vt))
print("\nIs A equal to the product of U, Sigma, and Vt? ", np.allclose(A, A_reconstructed))

# Rank 2 approximation of matrix A
rank_2_approximation = np.dot(U[:, :2], np.dot(np.diag(Sigma[:2]), Vt[:2, :]))
print("\nRank 2 approximation of matrix A:")
print(rank_2_approximation)

# Rank 3 approximation of matrix A
rank_3_approximation = np.dot(U[:, :3], np.dot(np.diag(Sigma[:3]), Vt[:3, :]))
print("\nRank 3 approximation of matrix A:")
print(rank_3_approximation)
```

# **Q4. Polynomial Plotting with Annotations:**

```
import numpy as np
import matplotlib.pyplot as plt

# To set the random seed using the last two digits of my roll number
np.random.seed(9)

coefficients = np.random.randint(-10, 11, size=6)

def polynomial(x, coeffs):
    return np.polyval(coeffs, x)

x_values = np.linspace(-5, 5, 400)

y_values = polynomial(x_values, coefficients)

plt.plot(x_values, y_values, label='Polynomial')

# Find the local maxima and minima
maxima_indices = (y_values[:-2] < y_values[1:-1]) & (y_values[1:-1] > y_values[2:])
minima_indices = (y_values[:-2] > y_values[1:-1]) & (y_values[1:-1] < y_values[2:])
maxima_x = x_values[1:-1][maxima_indices]
maxima_y = y_values[1:-1][maxima_indices]
minima_x = x_values[1:-1][minima_indices]
minima_y = y_values[1:-1][minima_indices]

# Annotate maxima and minima on the plot
for x, y in zip(maxima_x, maxima_y):
    plt.annotate('Max', xy=(x, y), xytext=(-20, 10), textcoords='offset points', arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))
for x, y in zip(minima_x, minima_y):
    plt.annotate('Min', xy=(x, y), xytext=(-20, -20), textcoords='offset points', arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of a 5th degree polynomial with Maxima and Minima')
plt.legend()
plt.grid(True)
plt.show()
```
